```python
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
```

```python
house_prices = pd.read_csv("data.csv")
house_prices.head()
```

```python
house_prices.drop(columns=["date", "street", "city", "statezip", "country"], inplace=True)
house_prices.head()
```

```python
X = house_prices.drop(columns="price")
X.head()
```

```python
y = house_prices.loc[:, "price"]
y.head()
```

```python
price_prediction = tf.keras.models.Sequential([
    tf.keras.Input(shape=12, name="Input layer"),
    tf.keras.layers.Dense(units=6, activation=tf.keras.activations.linear, name="layer2"),
    tf.keras.layers.Dense(units=24, activation=tf.keras.activations.linear, name="layer3"),
    tf.keras.layers.Dense(units=1, name="Output")
])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, train_size=0.8, shuffle=True, random_state=9)
```

```python
price_prediction.compile(loss=tf.keras.losses.MeanAbsoluteError(), optimizer="adam")
price_prediction.fit(X_train, y_train)
```

```python
y_test.shape
```

```python
y_pred = price_prediction.predict(X_test)
```

```python
y_pred
```

```python
price_prediction.summary()
```

```python
price_prediction.evaluate(X_test, y_test)
```
